# Spin models

```python
IBH = icarogw.wrappers.spinprior_IBH()
HBH = icarogw.wrappers.spinprior_HBH()
PBH = icarogw.wrappers.spinprior_PBH_smeared()
```

## IBH

```python
chi = np.linspace(0,1,1000)

print(IBH.population_parameters)
IBH.update(delta_0=0.5,delta_dot_0=0.1,chi_ibh_0=0.1, chi_dot_ibh_0=0.1)

m = np.ones_like(chi)*5.
cost = np.ones_like(chi)
pdf = np.sqrt(IBH.pdf(chi,chi,cost,cost,m,m))
plt.plot(chi,pdf,label=r'$5 M_{\odot}$')

m = np.ones_like(chi)*30.
cost = np.ones_like(chi)
pdf = np.sqrt(IBH.pdf(chi,chi,cost,cost,m,m))
plt.plot(chi,pdf,label=r'$30 M_{\odot}$')

m = np.ones_like(chi)*100.
cost = np.ones_like(chi)
pdf = np.sqrt(IBH.pdf(chi,chi,cost,cost,m,m))
plt.plot(chi,pdf,label=r'$100 M_{\odot}$')

plt.legend()
plt.xlabel(r'$\chi$')
plt.ylabel(r'$p(\chi_1,\cos t|\Lambda,m)$')
```

```python
cost = np.linspace(-1,1,1000)

print(IBH.population_parameters)
IBH.update(delta_0=0.2,delta_dot_0=0.0,chi_ibh_0=0.5, chi_dot_ibh_0=0.2)

m = np.ones_like(chi)*5.
chi = np.ones_like(chi)*0
pdf = np.sqrt(IBH.pdf(chi,chi,cost,cost,m,m))
plt.plot(cost,pdf,label=r'$5 M_{\odot}$')

m = np.ones_like(chi)*30.
chi = np.ones_like(chi)*0
pdf = np.sqrt(IBH.pdf(chi,chi,cost,cost,m,m))
plt.plot(cost,pdf,label=r'$30 M_{\odot}$')

m = np.ones_like(chi)*100.
chi = np.ones_like(chi)*0
pdf = np.sqrt(IBH.pdf(chi,chi,cost,cost,m,m))
plt.plot(cost,pdf,label=r'$100 M_{\odot}$')

plt.legend()
plt.xlabel(r'$\cos t$')
plt.ylabel(r'$p(\chi_1,\cos t|\Lambda,m)$')
```

# Rates

```python
cw = icarogw.wrappers.FlatLambdaCDM_wrap(zmax=100.)

IBH_sw = icarogw.wrappers.spinprior_IBH()
IBH_mw = icarogw.wrappers.massprior_PowerLaw()
IBH_rw = icarogw.wrappers.rateevolution_Madau()
rate_IBH = icarogw.rates.CBC_astro_rate_spin(cw,IBH_mw,IBH_rw,IBH_sw)

HBH_sw = icarogw.wrappers.spinprior_HBH()
HBH_mw = icarogw.wrappers.massprior_PowerLaw()
HBH_rw = icarogw.wrappers.rateevolution_Madau()
rate_HBH = icarogw.rates.CBC_astro_rate_spin(cw,HBH_mw,HBH_rw,HBH_sw)

PBH_sw = icarogw.wrappers.spinprior_PBH_smeared()
PBH_mw = icarogw.wrappers.massprior_PowerLaw()
PBH_rw = icarogw.wrappers.rateevolution_Madau()
rate_PBH = icarogw.rates.CBC_astro_rate_spin(cw,PBH_mw,PBH_rw,PBH_sw)

astro_rate = icarogw.rates.CBC_astro_rate_spin_3pops(rate_IBH,rate_HBH,rate_PBH,common_pop_parameters=['alpha','mmin','mmax','gamma','kappa','zp'])
```
